```python
import pandas as pd
from sklearn.model_selection import train_test_split

X_full = pd.read_csv('../../dataset/train.csv', index_col='Id')
X_test_full = pd.read_csv('../../dataset/test.csv', index_col='Id')

# obtain target and predictor
y = X_full.SalePrice
# features = ['LotArea','YearBuilt', '1stFlrSF', '2ndFlrSF',
#             'OverallQual','OverallCond', 'GrLivArea',
#             'FullBath','BedroomAbvGr','TotRmsAbvGrd',
#             'Fireplaces','WoodDeckSF','YrSold']
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 
            'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()

X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  train_size=0.8,
                                                  test_size=0.2,
                                                  random_state=0)
```

```python
X.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_4 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)
# model_5 = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)

models = [model_1, model_2, model_3, model_4]#, model_5]
```

```python
from sklearn.metrics import mean_absolute_error

def score_model(model, X_t=X_train, X_v=X_val, y_t=y_train, y_v=y_val):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
```

```python
best_model = model_4
fin_model = best_model.fit(X_train, y_train)
fin_model.fit(X, y)

preds_test = fin_model.predict(X_val)

output = pd.DataFrame({'Id': X_val.index,
                       'SalePrice': preds_test})
output.to_csv('../../dataset/submission_intermediate.csv', index=False)
```
